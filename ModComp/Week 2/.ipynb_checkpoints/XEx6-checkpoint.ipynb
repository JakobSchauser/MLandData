```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plots(X,Y,Z):
    fig, axs = plt.subplots(1,3,figsize=(14,4))
    
    axs[0].scatter(x = jitter(X),y = jitter(Y),alpha = 0.3)
    axs[0].set_title("Both")
    
    axs[1].scatter(x = jitter(X[Z==0]),y = jitter(Y[Z==0]),alpha = 0.3)
    axs[1].set_title("Z=0")

    axs[2].scatter(x = jitter(X[Z==1]),y = jitter(Y[Z==1]),alpha = 0.3)
    axs[2].set_title("Z=1")
    
    plt.show()
    
def jitter(values):
    return values + np.random.normal(0,0.3,values.shape)
```

```python
N = 1000

Z = np.round(np.random.random(N)).astype(int)
X = np.empty(N)
Y = np.empty(N)


X[Z==0] = np.random.poisson(1, N)[Z==0]
X[Z==1] = np.random.poisson(5, N)[Z==1] 


Y[Z==0] = np.random.poisson(1, N)[Z==0]
Y[Z==1] = np.random.poisson(5, N)[Z==1] 


plots(X,Y,Z)
```

```python
Z = np.round(np.random.random(N)).astype(int)
X = np.empty(N)
Y = np.empty(N)


X = np.random.poisson(5, N)
Y = np.random.poisson(5, N)

Z = (np.abs(X-Y) <= 2).astype(int)

plots(X,Y,Z)
```
