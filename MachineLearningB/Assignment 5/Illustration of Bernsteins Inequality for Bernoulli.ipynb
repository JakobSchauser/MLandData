```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
```

```python
exp = np.round(np.random.random((N,int(1e6))))
```

```python
hoeff = lambda x: np.minimum(1,np.exp(-2*N*x*x))

def bernstein(epsilon,mu):
    return np.minimum(1,np.exp(-N*epsilon*epsilon/(2*mu*(1-mu) + 2*0.5*epsilon/3)))
```

```python
exp= np.random.choice([0,1],size=(N,int(1e6)))

exp_c = np.sum(exp,axis = 0)/N

A = np.linspace(0.5,1,11)-0.000001

empirical = [np.sum(exp_c>=a)/len(exp_c) for a in A]



plt.plot(A, empirical,label='Empirical')
plt.plot(A,hoeff(A-0.5),label = 'Hoeffding')
plt.plot(A,bernstein(A-0.5,0.5),label = 'Bernstein')
plt.legend()
plt.xlabel("$\\alpha$")
plt.ylabel("$p(\mu\geq\\alpha)$")
plt.show()
```
