```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
```

```python
exp = np.round(np.random.random((N,int(1e6))))
```

```python
exp_c = np.sum(exp,axis = 0)/20

A = np.linspace(0.5,1,11)

empirical = [np.sum(exp_c>a)/len(exp_c) for a in A]
```

```python
hoeff = lambda x: np.minimum(1,np.exp(-2*N*x*x))

def bernstein(epsilon):
    return 2*np.exp(-N*epsilon*epsilon/(2*(1+epsilon/3)))
```

```python
plt.plot(A, empirical,label='Empirical')
plt.plot(A,hoeff(0.5-A),label = 'Hoeffding')
plt.plot(A,bernstein(A),label = 'Bernstein')
plt.legend()
plt.show()
```
