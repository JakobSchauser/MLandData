```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("white")

dat = pd.read_csv("ML-B 2021 - Final Exam Comments - studerende(1).csv")
dat.drop("Exam id", inplace=True, axis=1)
```

```python
good = (dat == "ok") | (dat == "OK") | (dat == 'NaN') | (dat == 'Nice.') | (dat == "rigorous derivation missing") | dat.isnull()
# good.head()
dat.head()
```

```python
y = np.mean(good)*100
s = np.argsort(y)

qs = [[g[1] == str(i+1) for g in good.columns[s]] for i in range(6)]


fig, axs = plt.subplots(1,1,figsize = (10,5))

axs.plot(good.columns[s],y[s], '.', markersize = 1)


cs = ["royalblue","darkorange","darkgreen"]


for i,q in enumerate(qs):
    axs.plot(good.columns[s][q],y[s][q], '.', markersize = 10,label = f"Q{i+1}",color = cs[i+1 == 5 or i+1 == 3])
    

# plt.legend()
plt.title("Spot the Yevgeny questions")
axs.set_ylabel("Percentage of positive comments")
sns.despine()
plt.show()
```
