```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets

import matplotlib.pyplot as plt
%matplotlib inline
```

Load handwritten digits data:

```python
digits = datasets.load_digits()
imshape = digits.images[0].shape  # The 2D image shape
X = digits.images.reshape(digits.images.shape[0],-1)  # Image
```

Do the PCA:

```python
pca = PCA()
pca.fit(X)
```

Find out if 10 components are enough to explain 80% of the variance:

```python
cumvar = np.cumsum(pca.explained_variance_ratio_)
plt.plot(cumvar)
print(cumvar[9])
```

Plot the eigenspectrum:

```python
plt.plot(pca.singular_values_**2)
plt.yscale("log")
```

Plot "eigendigits":

```python
fig, axs = plt.subplots(8,8,figsize = (14,14))
axs = axs.flatten()
for i in range(len(pca.components_)):
    axs[i].imshow(pca.components_[i].reshape(imshape),
              cmap=plt.cm.bone)
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```
